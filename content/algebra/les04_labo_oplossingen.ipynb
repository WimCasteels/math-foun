## Oefening 1: Stelsels Opstellen - Oplossingen

```python
# Opdracht 1a & 1b
print("Probleem 1: Appels en peren\n")

# 3a + 2p = 5
# 1a + 4p = 6
A1 = np.array([[3, 2],
               [1, 4]])
b1 = np.array([5, 6])

x1 = np.linalg.solve(A1, b1)
print(f"Prijs per appel: €{x1[0]:.2f}")
print(f"Prijs per peer: €{x1[1]:.2f}")
print(f"Verificatie: 3×{x1[0]:.2f} + 2×{x1[1]:.2f} = {3*x1[0] + 2*x1[1]:.2f} (moet 5 zijn)")
print()
```

```python
print("Probleem 2: Trein en wind\n")

# (s + w) * 2 = 300 → 2s + 2w = 300
# (s - w) * 3 = 300 → 3s - 3w = 300
A2 = np.array([[2, 2],
               [3, -3]])
b2 = np.array([300, 300])

x2 = np.linalg.solve(A2, b2)
print(f"Snelheid trein: {x2[0]:.1f} km/u")
print(f"Windsnelheid: {x2[1]:.1f} km/u")
print(f"Met wind: ({x2[0]:.1f} + {x2[1]:.1f}) × 2 = {(x2[0] + x2[1]) * 2:.0f} km")
print(f"Tegen wind: ({x2[0]:.1f} - {x2[1]:.1f}) × 3 = {(x2[0] - x2[1]) * 3:.0f} km")
```

```python
# Opdracht 1c
print("Probleem 3: Bakker\n")

# Onbekenden: prijs tarwe (t), rogge (r), meergranen (m)
# 10t + 5r + 8m = 67
# 8t + 10r + 6m = 70
# 12t + 8r + 10m = 88

A3 = np.array([[10, 5, 8],
               [8, 10, 6],
               [12, 8, 10]])
b3 = np.array([67, 70, 88])

x3 = np.linalg.solve(A3, b3)
print(f"Prijs tarwebrood: €{x3[0]:.2f}")
print(f"Prijs roggebrood: €{x3[1]:.2f}")
print(f"Prijs meergranenbrood: €{x3[2]:.2f}")
print()
print("Dit klopt met de gegeven prijzen: €2, €3, €4")
```

## Oefening 2: Stelsels Oplossen - Oplossingen

```python
# Opdracht 2a - Stelsel 1
print("Stelsel 1:\n")

A1 = np.array([[2, 1],
               [1, -1]])
b1 = np.array([5, 1])

x1 = np.linalg.solve(A1, b1)
print(f"Oplossing: x = {x1[0]}, y = {x1[1]}")
print(f"Verificatie: A @ x = {A1 @ x1}")
print(f"            b     = {b1}")
print()
```

```python
# Opdracht 2a - Stelsel 2
print("Stelsel 2:\n")

A2 = np.array([[1, 1, 1],
               [2, -1, 1],
               [1, 2, -1]])
b2 = np.array([6, 3, 3])

x2 = np.linalg.solve(A2, b2)
print(f"Oplossing: x = {x2[0]}, y = {x2[1]}, z = {x2[2]}")
print(f"Verificatie: A @ x = {A2 @ x2}")
print(f"            b     = {b2}")
```

```python
# Opdracht 2b - Oneindig veel oplossingen
print("Stelsel met oneindig veel oplossingen:\n")

A_inf = np.array([[1, 2],
                  [2, 4]])
b_inf = np.array([3, 6])

print(f"Matrix A:\n{A_inf}")
print(f"\nDeterminant: {np.linalg.det(A_inf)}")
print()
print("Rij 2 = 2 × Rij 1, dus de vergelijkingen beschrijven dezelfde lijn.")
print("Beide vergelijkingen zijn: x + 2y = 3")
print("Elke (x, y) op deze lijn is een oplossing.")
print("\nVoorbeelden: (3, 0), (1, 1), (-1, 2), ...")
```

```python
# Opdracht 2c - Geen oplossing
print("Stelsel zonder oplossing:\n")

A_no = np.array([[1, 2],
                 [2, 4]])
b_no = np.array([3, 8])

print(f"Matrix A:\n{A_no}")
print(f"b = {b_no}")
print(f"\nDeterminant: {np.linalg.det(A_no)}")
print()
print("De rijen zijn evenredig (rij 2 = 2 × rij 1),")
print("maar b[1] ≠ 2 × b[0] (8 ≠ 6).")
print("\nDit beschrijft twee parallelle lijnen die niet snijden.")
print("Lijn 1: x + 2y = 3")
print("Lijn 2: x + 2y = 4")
```

## Oefening 3: De Inverse Berekenen - Oplossingen

```python
# Opdracht 3a
print("Opdracht 3a: 2×2 inverse met de hand\n")

A = np.array([[4, 7],
              [2, 6]])

# Formule: A⁻¹ = (1/det) × [[d, -b], [-c, a]]
a, b = A[0, 0], A[0, 1]  # a=4, b=7
c, d = A[1, 0], A[1, 1]  # c=2, d=6

det = a*d - b*c  # 4×6 - 7×2 = 24 - 14 = 10

A_inv_manual = (1/det) * np.array([[d, -b],
                                    [-c, a]])

print(f"A = \n{A}")
print(f"\ndet(A) = {a}×{d} - {b}×{c} = {det}")
print(f"\nA⁻¹ (handmatig) = (1/{det}) × [[{d}, {-b}], [{-c}, {a}]]")
print(f"                = \n{A_inv_manual}")
print(f"\nA⁻¹ (NumPy) = \n{np.linalg.inv(A)}")
print(f"\nGelijk? {np.allclose(A_inv_manual, np.linalg.inv(A))}")
```

```python
# Opdracht 3b
print("Opdracht 3b: Verificatie A @ A⁻¹ = I\n")

A_inv = np.linalg.inv(A)

print("A @ A⁻¹ =")
print(A @ A_inv)
print()
print("A⁻¹ @ A =")
print(A_inv @ A)
print()
print(f"Beide zijn identiteitsmatrix? {np.allclose(A @ A_inv, np.eye(2))}")
```

```python
# Opdracht 3c
print("Opdracht 3c: 3×3 inverse\n")

B = np.array([[1, 2, 3],
              [0, 1, 4],
              [5, 6, 0]])

B_inv = np.linalg.inv(B)

print(f"B = \n{B}")
print(f"\nB⁻¹ = \n{B_inv}")
print(f"\nB @ B⁻¹ = \n{(B @ B_inv).round(10)}")
print(f"\nVerificatie: {np.allclose(B @ B_inv, np.eye(3))}")
```

```python
# Opdracht 3d
print("Opdracht 3d: Oplossen via inverse\n")

A = np.array([[4, 7],
              [2, 6]])
b = np.array([5, 3])

# Via inverse
x_inv = np.linalg.inv(A) @ b

# Via solve
x_solve = np.linalg.solve(A, b)

print(f"Oplossing via A⁻¹ @ b: {x_inv}")
print(f"Oplossing via solve:   {x_solve}")
print(f"\nGelijk? {np.allclose(x_inv, x_solve)}")
```

## Oefening 4: Determinanten - Oplossingen

```python
# Opdracht 4a
print("Opdracht 4a: Determinanten berekenen\n")

A = np.array([[3, 8], [4, 6]])
B = np.array([[1, 2], [3, 4]])
C = np.array([[2, 4], [1, 2]])

# Handmatig: det = ad - bc
det_A_hand = 3*6 - 8*4  # 18 - 32 = -14
det_B_hand = 1*4 - 2*3  # 4 - 6 = -2
det_C_hand = 2*2 - 4*1  # 4 - 4 = 0

print(f"Matrix A: det = 3×6 - 8×4 = {det_A_hand}")
print(f"  NumPy: {np.linalg.det(A):.1f}")
print()
print(f"Matrix B: det = 1×4 - 2×3 = {det_B_hand}")
print(f"  NumPy: {np.linalg.det(B):.1f}")
print()
print(f"Matrix C: det = 2×2 - 4×1 = {det_C_hand}")
print(f"  NumPy: {np.linalg.det(C):.1f}")
```

```python
# Opdracht 4b
print("Opdracht 4b: Inverteerbaarheid\n")

matrices = [('A', A), ('B', B), ('C', C)]

for name, M in matrices:
    det = np.linalg.det(M)
    invertible = abs(det) > 1e-10
    print(f"Matrix {name}: det = {det:.1f}, inverteerbaar: {invertible}")
    
    if invertible:
        print(f"  Inverse:\n{np.linalg.inv(M)}")
    else:
        print("  Geen inverse (singulier)")
    print()
```

```python
# Opdracht 4c
print("Opdracht 4c: Eigenschappen van determinanten\n")

np.random.seed(42)
A = np.random.randn(3, 3)
B = np.random.randn(3, 3)
c = 2

# 1. det(A × B) = det(A) × det(B)
det_AB = np.linalg.det(A @ B)
det_A_det_B = np.linalg.det(A) * np.linalg.det(B)
print(f"1. det(A @ B) = {det_AB:.6f}")
print(f"   det(A) × det(B) = {det_A_det_B:.6f}")
print(f"   Gelijk? {np.allclose(det_AB, det_A_det_B)}")
print()

# 2. det(Aᵀ) = det(A)
det_A = np.linalg.det(A)
det_AT = np.linalg.det(A.T)
print(f"2. det(A) = {det_A:.6f}")
print(f"   det(Aᵀ) = {det_AT:.6f}")
print(f"   Gelijk? {np.allclose(det_A, det_AT)}")
print()

# 3. det(cA) = c^n × det(A) voor n×n matrix
n = 3
det_cA = np.linalg.det(c * A)
cn_det_A = (c ** n) * np.linalg.det(A)
print(f"3. det({c}A) = {det_cA:.6f}")
print(f"   {c}³ × det(A) = {cn_det_A:.6f}")
print(f"   Gelijk? {np.allclose(det_cA, cn_det_A)}")
```

## Oefening 5: Geometrische Interpretatie - Oplossingen

```python
# Opdracht 5a
A = np.array([[3, -1],
              [1, 2]])
b = np.array([1, 12])

x_sol = np.linalg.solve(A, b)

# Plot de lijnen
x_range = np.linspace(-2, 8, 100)

# 3x - y = 1 → y = 3x - 1
y1 = 3*x_range - 1

# x + 2y = 12 → y = (12 - x) / 2
y2 = (12 - x_range) / 2

plt.figure(figsize=(10, 6))
plt.plot(x_range, y1, 'b-', linewidth=2, label='3x - y = 1')
plt.plot(x_range, y2, 'r-', linewidth=2, label='x + 2y = 12')
plt.plot(x_sol[0], x_sol[1], 'go', markersize=15, label=f'Snijpunt ({x_sol[0]:.1f}, {x_sol[1]:.1f})')

plt.xlim(-2, 8)
plt.ylim(-5, 10)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Stelsel als snijpunt van twee lijnen', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()

print(f"Oplossing: x = {x_sol[0]}, y = {x_sol[1]}")
```

```python
# Opdracht 5b
M1 = np.array([[2, 0], [0, 3]])
M2 = np.array([[1, 1], [0, 1]])
M3 = np.array([[0.5, 0], [0, 0.5]])

matrices = [('M1 (schaling)', M1), ('M2 (shear)', M2), ('M3 (krimp)', M3)]

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

square = np.array([[0, 1, 1, 0, 0],
                   [0, 0, 1, 1, 0]])

for ax, (name, M) in zip(axes, matrices):
    transformed = M @ square
    det = np.linalg.det(M)
    
    ax.fill(square[0], square[1], alpha=0.3, color='blue', label='Origineel (opp=1)')
    ax.plot(square[0], square[1], 'b-', linewidth=2)
    ax.fill(transformed[0], transformed[1], alpha=0.3, color='red', label=f'Getransformeerd (opp={abs(det):.2f})')
    ax.plot(transformed[0], transformed[1], 'r-', linewidth=2)
    
    ax.set_xlim(-0.5, 3)
    ax.set_ylim(-0.5, 3.5)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=9)
    ax.set_title(f'{name}\ndet = {det:.2f}', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Opdracht 5c
S = np.array([[1, 2],
              [2, 4]])

# Grid van punten
x = np.linspace(-1, 1, 11)
y = np.linspace(-1, 1, 11)
X_grid, Y_grid = np.meshgrid(x, y)
points = np.vstack([X_grid.ravel(), Y_grid.ravel()])

# Transformeer
transformed = S @ points

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].scatter(points[0], points[1], c='blue', alpha=0.6)
axes[0].set_xlim(-3, 3)
axes[0].set_ylim(-3, 6)
axes[0].set_aspect('equal')
axes[0].grid(True, alpha=0.3)
axes[0].set_title('Origineel grid', fontsize=12)

axes[1].scatter(transformed[0], transformed[1], c='red', alpha=0.6)
axes[1].set_xlim(-3, 3)
axes[1].set_ylim(-3, 6)
axes[1].set_aspect('equal')
axes[1].grid(True, alpha=0.3)
axes[1].set_title(f'Getransformeerd (det = {np.linalg.det(S):.1f})\nAlle punten op één lijn!', fontsize=12)

plt.tight_layout()
plt.show()

print("De singuliere matrix 'plat' het 2D vlak tot een 1D lijn.")
```

## Oefening 6: Lineaire Regressie from Scratch - Oplossingen

```python
# Opdracht 6a
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2.1, 3.9, 6.2, 7.8, 10.1])

# Design matrix
X = np.column_stack([np.ones(len(x_data)), x_data])
print("Design matrix X:")
print(X)
print()

# Normal equations
XtX = X.T @ X
Xty = X.T @ y_data

print(f"XᵀX = \n{XtX}")
print(f"\nXᵀy = {Xty}")
print()

# Oplossen
w = np.linalg.solve(XtX, Xty)
print(f"w = [intercept, slope] = {w}")
print(f"\nLijn: y = {w[0]:.4f} + {w[1]:.4f}x")
```

```python
# Visualisatie
y_pred = X @ w

plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_data, s=100, c='blue', label='Data')
plt.plot(x_data, y_pred, 'r-', linewidth=2, label=f'Fit: y = {w[0]:.2f} + {w[1]:.2f}x')

# Residuals
for i in range(len(x_data)):
    plt.plot([x_data[i], x_data[i]], [y_data[i], y_pred[i]], 'r--', alpha=0.5)

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Lineaire regressie via normal equations', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Opdracht 6b
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_data.reshape(-1, 1), y_data)

print("Vergelijking met sklearn:")
print(f"  Onze implementatie: intercept = {w[0]:.6f}, slope = {w[1]:.6f}")
print(f"  sklearn:            intercept = {model.intercept_:.6f}, slope = {model.coef_[0]:.6f}")
print(f"\nIdentiek? {np.allclose(w, [model.intercept_, model.coef_[0]])}")
```

```python
# Opdracht 6c - Polynomiale regressie
x_poly = np.array([0, 1, 2, 3, 4, 5])
y_poly = np.array([1, 2.5, 6, 11, 18, 27])

# Design matrix voor graad 2: [1, x, x²]
X_poly = np.column_stack([np.ones(len(x_poly)), x_poly, x_poly**2])
print("Design matrix voor polynomiale regressie:")
print(X_poly)
print()

# Oplossen
w_poly = np.linalg.solve(X_poly.T @ X_poly, X_poly.T @ y_poly)
print(f"Coëfficiënten [a, b, c]: {w_poly}")
print(f"\nPolynoom: y = {w_poly[0]:.4f} + {w_poly[1]:.4f}x + {w_poly[2]:.4f}x²")

# Visualisatie
x_smooth = np.linspace(0, 5, 100)
y_smooth = w_poly[0] + w_poly[1]*x_smooth + w_poly[2]*x_smooth**2

plt.figure(figsize=(10, 6))
plt.scatter(x_poly, y_poly, s=100, c='blue', label='Data')
plt.plot(x_smooth, y_smooth, 'r-', linewidth=2, 
         label=f'Fit: y = {w_poly[0]:.2f} + {w_poly[1]:.2f}x + {w_poly[2]:.2f}x²')
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Polynomiale regressie (graad 2)', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.show()
```

## Oefening 7: Pseudo-inverse - Oplossingen

```python
# Opdracht 7a
A = np.array([[1, 1],
              [2, 1],
              [1, 2],
              [2, 2]])
b = np.array([2, 3, 3, 5])

print(f"Overgedetermineerd systeem:")
print(f"A (shape {A.shape}):\n{A}")
print(f"\nb = {b}")
print()

# Pseudo-inverse
A_pinv = np.linalg.pinv(A)
x_pinv = A_pinv @ b

print(f"Oplossing via pseudo-inverse: x = {x_pinv}")
```

```python
# Opdracht 7b
residuals = A @ x_pinv - b
total_squared_error = np.sum(residuals**2)

print("Residuals (A @ x - b):")
for i, r in enumerate(residuals):
    print(f"  Vergelijking {i+1}: {r:.4f}")
print()
print(f"Totale kwadratische fout: {total_squared_error:.6f}")
```

```python
# Opdracht 7c
x_lstsq, residuals_lstsq, rank, s = np.linalg.lstsq(A, b, rcond=None)

print(f"Oplossing via pinv:  {x_pinv}")
print(f"Oplossing via lstsq: {x_lstsq}")
print(f"\nGelijk? {np.allclose(x_pinv, x_lstsq)}")
```

## Oefening 8: Inverse en Transformaties - Oplossingen

```python
# Opdracht 8a
theta = np.radians(30)
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta), np.cos(theta)]])

R_inv = np.linalg.inv(R)

# Rotatie van -30 graden
theta_neg = np.radians(-30)
R_neg = np.array([[np.cos(theta_neg), -np.sin(theta_neg)],
                  [np.sin(theta_neg), np.cos(theta_neg)]])

print("Rotatiematrix R (30°):")
print(R)
print()
print("Inverse R⁻¹:")
print(R_inv)
print()
print("Rotatiematrix R(-30°):")
print(R_neg)
print()
print(f"R⁻¹ = R(-30°)? {np.allclose(R_inv, R_neg)}")
print("\nDe inverse van een rotatie is de rotatie in de tegengestelde richting!")
```

```python
# Opdracht 8b
triangle = np.array([[0, 1, 0.5, 0],
                     [0, 0, 1, 0]])

# Transformaties
after_R = R @ triangle
after_R_Rinv = R_inv @ after_R

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Origineel
axes[0].fill(triangle[0], triangle[1], alpha=0.3, color='blue')
axes[0].plot(triangle[0], triangle[1], 'b-', linewidth=2)
axes[0].set_title('Origineel', fontsize=12)

# Na R
axes[1].fill(triangle[0], triangle[1], alpha=0.2, color='blue')
axes[1].fill(after_R[0], after_R[1], alpha=0.3, color='red')
axes[1].plot(after_R[0], after_R[1], 'r-', linewidth=2)
axes[1].set_title('Na rotatie R (30°)', fontsize=12)

# Na R_inv
axes[2].fill(after_R[0], after_R[1], alpha=0.2, color='red')
axes[2].fill(after_R_Rinv[0], after_R_Rinv[1], alpha=0.3, color='green')
axes[2].plot(after_R_Rinv[0], after_R_Rinv[1], 'g-', linewidth=2)
axes[2].set_title('Na R⁻¹ (terug naar origineel)', fontsize=12)

for ax in axes:
    ax.set_xlim(-0.5, 1.5)
    ax.set_ylim(-0.5, 1.5)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Terug bij origineel? {np.allclose(triangle, after_R_Rinv)}")
```

```python
# Opdracht 8c (Bonus)
print("Opdracht 8c: R⁻¹ = Rᵀ voor rotatiematrices\n")

print("R⁻¹:")
print(R_inv)
print()
print("Rᵀ:")
print(R.T)
print()
print(f"R⁻¹ = Rᵀ? {np.allclose(R_inv, R.T)}")
print()
print("Dit geldt voor alle rotatiematrices!")
print("Rotatiematrices zijn 'orthogonaal': R @ Rᵀ = I")
```

## Bonusoefening: Numerieke Instabiliteit - Oplossing

```python
# Bonus: Bijna-singuliere matrices
print("Numerieke instabiliteit bij bijna-singuliere matrices\n")

perturbations = [0.1, 0.01, 0.001, 0.0001, 0.00001]

for eps in perturbations:
    A = np.array([[1, 2], [1, 2 + eps]])
    det = np.linalg.det(A)
    
    A_inv = np.linalg.inv(A)
    product = A @ A_inv
    error = np.max(np.abs(product - np.eye(2)))
    
    print(f"ε = {eps}")
    print(f"  det(A) = {det:.2e}")
    print(f"  max|A @ A⁻¹ - I| = {error:.2e}")
    print()

print("Conclusie: hoe kleiner de determinant, hoe groter de numerieke fouten!")
```
